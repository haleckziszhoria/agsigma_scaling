```python
import random
  

modelo = [1,1,0,0,0,1,1,0,1] #Objetivo a alcanzar
largo = 9 #La longitud del material genetico de cada individuo
num = 10 #La cantidad de individuos que habra en la poblacion
cant_indiv = 3 #Cuantos individuos se seleccionan para reproduccion. Necesariamente mayor que 2
mutacion_prob = 0.2 #La probabilidad de que un individuo mute
  
print("\n\nModelo: %s\n"%(modelo)) #Mostrar el modelo, con un poco de espaciado
```

```python
def individual(min, max):
    # Crea un individuo con valores aleatorios
    return[random.randint(min, max) for i in range(largo)]
```

```python
def crearPoblacion():
    #Crea una poblacion de 10 individuos
    return [individual(0,1) for i in range(num)]
```

```python
def calcularFitness(individual,avg):
    
    fitness = 0
    for i in range(len(individual)):
        if individual[i] == modelo[i]:
            fitness += 1
    #con sigma scaling
    c = random.randint(1, 5)
    sigma = fitness + (avg*c)
    fitness = fitness * sigma
    return fitness
```

```python
def selection_and_reproduction(poblacion):
    temp = random.randint(0,1)
    puntuados = [ (calcularFitness(i,temp), i) for i in poblacion] 
    puntuados = [i[1] for i in sorted(puntuados)] 
    poblacion = puntuados
  
  
  
    selected =  puntuados[(len(puntuados)-cant_indiv):] 
  
  
  
    
    for i in range(len(poblacion)-cant_indiv):
        punto = random.randint(1,largo-1) 
        padre = random.sample(selected, 2) 
          
        poblacion[i][:punto] = padre[0][:punto] 
        poblacion[i][punto:] = padre[1][punto:]
        print("\nPadre:",padre,"\nPunto de Cruce: ",punto,"\nIteración Nro:",i,"Poblacion Actual:\n",population)
  
    return poblacion 
```

```python
def mutation(poblacion):
    """
        Se mutan los individuos al azar. Sin la mutacion de nuevos genes nunca podria
        alcanzarse la solucion.
    """
    print("=================Mutacion========================")
    for i in range(len(poblacion)-cant_indiv):
        if random.random() <= mutacion_prob: #Cada individuo de la poblacion (menos los padres) tienen una probabilidad de mutar
            punto = random.randint(0,largo-1) #Se elgie un punto al azar
            nuevo_valor = random.randint(0,1) #y un nuevo valor para este punto
            print("Punto de Mutacion: ", punto, "Nuevo valor: ",nuevo_valor)
            #Es importante mirar que el nuevo valor no sea igual al viejo
            while nuevo_valor == poblacion[i][punto]:
                nuevo_valor = random.randint(0,1)
  
            #Se aplica la mutacion
            poblacion[i][punto] = nuevo_valor
  
    return poblacion
```

```python
population = crearPoblacion()
print("Poblacion Inicial:\n%s"%(population)) 
    
#Se evoluciona la poblacion
for i in range(10):
    print("=====================================Generacion Nro: ",i,"============================")
    population = selection_and_reproduction(population)
    population = mutation(population)
  
print("\nPoblacion Final:\n%s"%(population)) #Se muestra la poblacion evolucionada
print("\n\n")
```
